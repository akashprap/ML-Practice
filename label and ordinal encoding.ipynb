## Label and Ordinal Encoding 
### for categorical data having ranks associated -----> OrdinalEncoder
### for categorical data that don't have rank associated ------> LabelEncoder 
LabelEncoder does not create new features for each unique category,rather it gives rank to those unique features


```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset('titanic')
```

```python
df.sample(5)
```

```python
df['class'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
ord_encoder=OrdinalEncoder(categories=[['First','Second','Third']])
lab_encoder=LabelEncoder()
```

```python
lab_encoder.fit_transform(df['embark_town'])
```

```python
df['class_encoded']=ord_encoder.fit_transform(df[['class']])
```
