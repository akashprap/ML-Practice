## Smote is used for upscaling and downscaling imbalanced dataset

```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
# X independent feature
# y dependent feature
X,y=make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=1,n_redundant=0)
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
df=pd.concat([df1,df2],axis=1)
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
# for performing SMOTE installed "imblearn" module
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
X,y=oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
df1=pd.DataFrame(X,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df1,df2],axis=1)
```
